```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from xgboost import XGBClassifier
```

```python
url = "https://raw.githubusercontent.com/nsethi31/Kaggle-Data-Credit-Card-Fraud-Detection/master/creditcard.csv"
df = pd.read_csv(url)
```

```python
df
```

```python
df=df.sample(frac=1,random_state=42)
```

```python
x=df.drop('Class',axis=1)
y=df['Class']
```

```python
from sklearn.preprocessing import StandardScaler
x_scaler=StandardScaler()
x=x_scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
pre=model.predict(x_test)
```

```python
accuracy_score(y_test,pre)
```

```python
from xgboost import XGBClassifier
model_xgb=XGBClassifier(use_label_encoder=False,evaluation_metric='logloss',n_estimators=100,random_state=42)
model_xgb.fit(x_train,y_train)
```

```python
pre_xgb=model_xgb.predict(x_test)
```

```python
accuracy_score(y_test,pre_xgb)
```

```python
print('The classification report for the Random Forest Classifier is as follows:\n',classification_report(y_test,pre))
```

```python
def predict_transaction(TransactionData):
  transaction_scaled=x_scaler.transform([TransactionData])
  rf_prediction=model.predict(transaction_scaled)[0]
  xgb_prediction=model_xgb.predict(transaction_scaled)[0]
  print('Random Forest Prediction is',rf_prediction,'and','The XGBOOSTING Prediction is:',xgb_prediction)
```

```python
exp_transaction=x_test[1]
prediction=predict_transaction(exp_transaction)
```

```python
print('The real Time Transaction prediction is',prediction)
```
